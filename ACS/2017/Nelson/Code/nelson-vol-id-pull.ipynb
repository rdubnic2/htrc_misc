```python
ny_org_pull_list = []

for i in ny_org_df.index:
    ben_f = ny_org_df['pull for Ben (focused)'][i]
    ben_b = ny_org_df['pull for Ben (broad)'][i]
    name = ny_org_df['ny org'][i]
    nan_test_f = pd.isnull(ben_f)
    nan_test_b = pd.isnull(ben_b)
    # print(ben)
    if nan_test_f == False or nan_test_b == False:
        ny_org_pull_list.append(name)

print(len(ny_org_pull_list))
```

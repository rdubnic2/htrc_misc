```python
import numpy as np
import pandas as pd
import glob
from pathlib import Path
```

```python
vector_directory = glob.glob('/Users/rdubnic2/Desktop/data/sample-numpy-vectors/*.np*')
for v in vector_directory:
    v = Path(v)
```

```python
expected_data_type = "<class 'numpy.float32'>"
outfile_name = 'enderle_vector_verify.txt'

with open(outfile_name,'a') as f:
    for vector in vector_directory:
        split_v = vector.split('/')
        vector_name = split_v[-1]
        # print(vector_name)
        
        try:
            loaded_vector = np.load(vector, allow_pickle=True)
        except:
            print(f"There was a problem loading vector filepath {vector} due to failure of np.load", file=f)
            print(f"There was a problem loading vector filepath {vector} due to failure of np.load")
        
        test = type(loaded_vector)
        shape = loaded_vector.shape
        data_type = loaded_vector.dtype.type
        str_data_type = str(data_type)
        vec_sum = np.sum(loaded_vector)
        # print(vec_sum)
        vec_has_nan = np.isnan(vec_sum)
        # print(vec_has_nan)
        
        if str_data_type != expected_data_type:
                print(f"There was a problem loading vector filepath {vector} due to not matching expected data type", file=f)
                print(f"There was a problem loading vector filepath {vector} due to not matching expected data type")
        
        elif vec_has_nan == True:
            print(f"There was a problem loading vector filepath {vector} due to NaN element in array", file=f)
            print(f"There was a problem loading vector filepath {vector} due to NaN element in array")
        
        else:
            print(f"Successfully loaded {vector_name}, a {test} of shape {shape}", file=f)
```
